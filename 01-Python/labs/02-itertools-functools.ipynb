### Snippet 4 -- partial — positional vs keyword  
> Mi predicción: La primera parte , le pasas base y exponente y te da la potencia. Partial es como que te crea una funcion y se la anade a (en este caso) square_kw y square_pos..ya que no veo que tengan para recibir args, pero se les esta pasando en el print valores. Osea square_kw(3) -> partial(3, exp=2)

```python
from functools import partial

def power(base, exp):
    return base ** exp

square_kw = partial(power, exp=2)
square_pos = partial(power, 2)

print(square_kw(3))
print(square_pos(3))
```

> Output: No se que carajos paso aca. Por que el mismo valor(3) me retorna distintos resultados. Algo tiene que ver con exp. Pero no se que.

> Cuando llamás square_kw(3):  
exp ya vale 2 (bindeado).  
El 3 rellena el siguiente arg disponible → base = 3.  
Resultado: power(base=3, exp=2) = 3² = 9.  
Cuando llamás square_pos(3):   
base ya vale 2 (bindeado posicionalmente como primer arg).  
El 3 rellena el siguiente arg disponible → exp = 3.  
Resultado: power(base=2, exp=3) = 2³ = 8.  
Por eso el mismo input (3) da resultados distintos. El partial con keyword vs positional bindea cosas distintas. Lección operativa: siempre que uses partial, preferí keyword args (exp=2). Es más explícito y menos confuso.

## 4. reduce(función, lista)  
Toma una lista y la "colapsa a un solo valor" aplicando la función pair-wise:

```python
reduce(lambda a, b: a + b, [1,2,3,4])
```

Paso a paso:  
(1+2) → 3  
(3+3) → 6  
(6+4) → 10  
Resultado final: 10  
Es como ir acumulando: tomá los primeros 2 elementos, aplicales la función, el resultado se vuelve a para la siguiente iteración con el próximo elemento como b

## 5. partial(función, args)  
Pre-llena algunos argumentos de una función y te devuelve una nueva función con los argumentos restantes pendientes:

```python
def power(base, exp): return base ** exp

cuadrado = partial(power, exp=2)  # exp ya está fijo en 2
cuadrado(5)  # equivalente a power(5, 2) = 25
cuadrado(3)  # equivalente a power(3, 2) = 9
```

Es útil cuando tenés una función general y querés "especializarla" sin reescribirla.

## For ML
1. `zip`: iterar features y labels juntos. for x, y in zip(features, labels): — patrón básico de loops de training.
2. `enumerate`: loggear progreso con epoch number durante entrenamiento. for epoch, batch in enumerate(dataloader, start=1): print(f"Epoch {epoch}...")
3. `reduce`: raramente lo vas a usar tal cual — sum(), max(), min() ya están — pero ENTENDER reduce te ayuda a leer código de funciones de pérdida (loss functions) que acumulan valores.
4. `partial`: crear funciones de pérdida o métricas con hyperparameters fijos. loss_fn = partial(weighted_loss, weight=0.7) — útil cuando la API de un framework te pide una función con cierta firma específica.
5. `cycle + islice`: infinite training data loop. for batch in islice(cycle(dataloader), num_steps): — patrón clásico para entrenar exactamente N steps sin pensar en epochs.
