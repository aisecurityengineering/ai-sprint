### Snippet 1 — Mutable default argument
> Mi predicción: Un argumento que por default puede mutar en cualquier tipo de valor (str, int, float)

```python
#1. Mutable default
def add(item, lst=[]):
    lst.append(item)
    return lst
    
print(add(1))
print(add(2))
```

> Output: Dos arrays, el primer array recibe un 1 y al segundo array se le suma el 2.
Como difiere de mi prediccion: No se si entendi bien, el mutable argument for default aqui es el lst=[]. Es el que muta en base al valor de item.

> GOTCHA: lst=[] se evalua una sola vez, cuando se define la funcion. Esa misma lista se rehusa entre llamadas. Por eso add(1) → [1], y add(2) → [1, 2] (la lista persiste).  
Pro rule: nunca usar mutables como default

Mas info: https://docs.python-guide.org/writing/gotchas/

```python
# Correct pattern
def add(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add(1))
print(add(2))
```

### Snippet 2 — Identity vs equality (`is` vs `==`)
> Mi predicción: is es el tipo y '==' si son de igual tipo

```python
#2. Identity vs equality
a = 256
b = 256
c = 257
d = 257
print(a is b, c is d)
```

> Output: a y b son 256 y al compararse dan True. c y d son 257 y al compararse dan False.
Como difiere de mi prediccion: Muchisimo, ni siquiera entiendo que es lo que pasa.

> Gotcha: `is` compara identidad (son el mismo objeto en memoria?) CPython pre-crea objetos enteros del -5 al 256 y los comparte entre variables — eso se llama integer interning. Por eso 256 is 256 es True (mismo objeto en memoria), 257 is 257 es False (objetos distintos). Regla: nunca uses is para comparar valores. Usalo solo para is None, is True, is False.

### Snippet 3 — Copy semantics (`copy.copy` con anidados)
> Mi predicción: Copiar el contenido anidado dentro de otro contenido copiado.

```python
#3. Copy semantics
import copy
x = [[1,2], [3,4]]
y = copy.copy(x)
y[0].append(99)
print(x)
```

> Output: se copia el valor de x en y. Al parecer y ahora apunta al mismo lugar en memoria que x.
Como difiere de mi prediccion: Bastante, lo unico que no sabia era que tabien copia el lugar de la memoria (cosa que creo que es lo que hace)

> GOTCHA:  El término es shallow copy vs deep copy. copy.copy(x) crea una nueva lista exterior pero las interiores siguen siendo las MISMAS referencias. Modificar y[0] afecta a x[0]. Para copia profunda total: copy.deepcopy(x)

### Snippet 4 — Generator vs list (memoria)
> Mi predicción: Generador era el texto dentro de """ """ para decir que es lo que hace el codigo. List, es una lista con valores los cuales, no recuerdo que caracteristica debian de tener.

```python
#4. Generatos vs Lists
import sys
a = [i for i in range(1000)] # List
b = (i for i in range(1000)) # Generator
print(sys.getsizeof(a), sys.getsizeof(b))
print(a)
print(b)
```

> Output: Se rellenan los valores a y b con los numeros del 0 al 999. 
Como difiere de mi prediccion: Muchisimo. No tenia nada que ver.

> GOTCHA: La diferencia es brutal. List guarda los 1000 valores en RAM. Generator guarda solo la receta para producirlos bajo demanda, uno por uno. Para datasets grandes (todo lo que viene en ML), generators son la diferencia entre "carga el dataset entero a memoria y muere" y "procesa fila por fila al infinito".

### Snippet 5 — for-else
> Mi predicción: Es un loop for-else el cual itera realizando acciones dependiendo los valores.

```python
#5. For-else
for i in range(3):
    if i == 5:
        break
else:
    print("done")
```

> Output real: done  
Cómo difiere de mi predicción:En nada, es el que mas claro tenia de todos.

> GOTCHA: No, no lo tenías claro — solo se vio simple porque tu predicción era vaga. El mecanismo real: else ejecuta solo si el for completó sin pasar por break. range(3) produce 0, 1, 2 — nunca llega a 5, nunca hace break, por eso "done" se imprime. Si cambiás a for i in range(10) con mismo if i == 5: break, el "done" NO se imprime. Probalo. Es el feature más anti-intuitivo de Python.

```python
#5. For-else
for i in range(10):
    if i == 5:
        break
else:
    print("done")
# Esperado: no imprime "done"
```

```python
for i in range(3):
    pass
else:
    print("done")
```

> El else, se ejecuta solo si el for termina sin break
