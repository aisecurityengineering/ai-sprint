### Ejercicio 2 — Reference vs copy
Concepto: este ES el bug que mata más pipelines de ML. 

````python
#2.1 Reference (lo que ya existe)
a = [1,2,3]
b = a
a.append(4)
# Prediccion: que imprime b?
# b = [1,2,3,4]
print(b)
````

````python
# 2.2 Shallow copy
import copy
a = [1,2, [3,4]]
b = a.copy()
a.append(5)
# Prediccion: b cambia?
# b = [1,2, [3,4]]
print(b)
print(a)
````

````python
a[2].append(99)
# Prediccion: b cambia ahora?
# b = [1,2, [3,4,99]]
print(b)
````

````python
# 2.3 Deep Copy
a = [1,2,[3,4]]
b = copy.deepcopy(a)
a[2].append(99)
# Prediccion cambia b?
print(b)
````

````python
# Formas de copiar
b = a # Apunta al mismo espacio en memoria
b = a.copy() # Apunta al mismo espacio en memoria
b = copy.deepcopy(a) # No apunta al mismo espacio en memoria
````

### Ejercicio 3 — Type checking y duck typing

````python
# 3.1 Dada esta lista mixta
mixed = [1, "2", 3.0, True, None, [1,2], (1, 2), {"key":"val"}]

# Escribí una función que cuente cuántos elementos hay de cada tipo principal:
# int, str, float, bool, None, list, tuple, dict
# Tip: cuidado con bool — en Python, True ES INSTANCIA de int. Pensá si querés 
# que True cuente como int o como bool separado.

def count_types(items):
    results = {
        "integers" : 0,
        "strings" : 0,
        "float_numbers" : 0,
        "booleanos" : 0,
        "nones" : 0,
        "lists" : 0,
        "tuples" : 0,
        "dictionaries" : 0
    }

    # completar usando isinstance
    for i in items:
        if(isinstance(i,bool)):
            results["booleanos"] += 1
        elif isinstance(i, str):
            results["strings"] += 1
        elif(isinstance(i,float)):
            results["float_numbers"] += 1
        elif(isinstance(i, int)):
            results["integers"] += 1
        elif(isinstance(i, type(None))):
            results["nones"] += 1
        elif(isinstance(i, list)):
            results["lists"] += 1
        elif(isinstance(i, tuple)):
            results["tuples"] += 1
        elif(isinstance(i, dict)):
            results["dictionaries"] += 1

    return results
      
res = count_types(mixed)
print(res)
````

````python
# 3.2 Usando la funcion isiterable que aprendiste, separa los iterables
# de los no iterables (excepto strings -- strings SON iterables pero queremos tratarlas distinto)
def split_iterables(items):
    """Return (iterables, non_iterables, strings)"""
    iterables = []
    noniterables = []

    for i in items:
        
        if(isinstance(i, str)):
            noniterables.append(i)
        else:
            try: 
              if(iter(i)):
                  iterables.append(i)
            except:
                  noniterables.append(i)
    return [iterables, noniterables]

res = split_iterables(mixed)
print(res[0]) # iterables
print(res[1]) # non-iterables
# Probar con 'mixed'
````

````python
# 4.3 Receipt printer - dado:
items = [("Apple", 3, 0.50), ("Bread", 1, 2.30), ("Coffee", 2, 4.99)]

# Imprimí un recibo con columnas alineadas:
# Apple       3 x $0.50  =  $1.50
# Bread       1 x $2.30  =  $2.30
# Coffee      2 x $4.99  =  $9.98
# ----------------------------------
# TOTAL                  = $13.78
# Usá f-strings con specifiers de ancho fijo y decimales.

def make_receipt(items):
    total = 0

    for name, qty, price in items:
        subtotal = qty * price
        total += subtotal

        print(f"{name:<10} {qty:>2} x ${price:>5.2f}  =  ${subtotal:>6.2f}")

    print("-" * 34)
    print(f"{'TOTAL':<20}     = ${total:>6.2f}")

make_receipt(items)
````

### Ejercicio 5 — Control flow + range

````python
# 5.1 FizzBuzz clasico - del 1 al 30:
# Si el numero es multiplo de 3 -> "Fizz"
# Si es multiplo de 5 -> "Buzz"
# Si es multiplo de ambos -> "FizzBuzz"
# Sino -> el numero
for i in range(1, 31):
    #complete
    if (i % 3 == 0 and i % 5 == 0):
        print("FizzBuzz")
    elif (i % 3 == 0):
        print("Fizz")
    elif (i % 5 == 0):
        print("Buzz")
    else:
        print(i)
````

````python
# 5.2 Encontrar primos < 50 usando break/continue
# Hint: para cada n del 2 al 50, chequeás divisibilidad por 2 hasta n-1
# Si encontrás divisor → no es primo (break)
# Si recorriste todo sin encontrar divisor → es primo
# Usá el patrón for-else que aprendiste hace 2 días

for n in range(2,50):
    for i in range(2, n):
        if n % i == 0:
            break
    else:
        print(f"{n} es primo")
````

````python
# 5.3 Bonus — ¿cuál de estas dos formas es más rápida?
%timeit sum(i for i in range(1000) if i % 3 == 0 or i % 5 == 0)
%timeit sum([i for i in range(1000) if i % 3 == 0 or i % 5 == 0])
# Anotá cuál ganó y por qué (generator vs list — concepto del Día 3 de gotchas)
# Gana generator ya que no guarda en RAM los valores
# List guarda los 1000 valores en RAM. Generator guarda solo la receta para producirlos bajo demanda, uno por uno
````
