## Expressing Conditional Logic as Array 

The `numpy.where` function is a vectorized version of the ternary expression `x if condition else y`. Suppose we had a Boolean array and two arrays of values

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
# Suppose we wanted to take a value from xarr whenever the corresponding value in cond is True, 
# and otherwise take the value from yarr. 
# A list comprehension doing this might look like:
result = [(x if c else y) for x,y,c in zip(xarr, yarr, cond)]
result
```

```python
# Better way ...EASY!!
result = np.where(cond, xarr, yarr)
result
```

### Ejercicio 1 — Vectorization speed test real

```python
import numpy as np
# Tarea: clasificar 1 millon de "HTTP Status Code" en categorias:
# 2xx = success, 3xx = redirect, 4xx = client error, 5xx = server error 
rng = np.random.default_rng(seed=42)
status_codes = rng.integers(200, 600, size=1_000_000)
status_codes
```

```python
#1.1 Version python pura
def classify_python(codes):
    result = []
    for c in codes:
        if 200 <= c < 300:
            result.append("success")
        elif 300 <= c < 400:
            result.append("redirect")
        elif 400 <= c < 500:
            result.append("client_error")
        else:
            result.append("server_error")
    return result

# 1.2 Versión NumPy con np.where anidado o np.select — completá:
def classify_numpy(codes):
    conditions = [
        (codes >= 200) & (codes < 300),
        (codes >= 300) & (codes < 400),
        (codes >= 400) & (codes < 500)
    ]

    options = ["success", "redirect", "client_error"]

    return np.select(conditions, options, default="server_error")

res = classify_numpy(status_codes)
# 1.3 Compará tiempos
%timeit classify_python(status_codes)
%timeit classify_numpy(status_codes)
# Predicción: ¿cuántas veces más rápido es NumPy?
```

### Ejercicio 4 — Reshaping y broadcasting para batches 

```python
# Simulemos un "batch" de 32 imágenes RGB de 64x64 píxeles
# Shape: (32, 64, 64, 3) — batch_size, height, width, channels
rng = np.random.default_rng(seed=42)
batch = rng.integers(0, 256, size=(32, 64, 64, 3), dtype=np.uint8)
```

```python
# 4.3 Normalizar el batch entero a rango [0, 1]
batch_norm = batch / 255.0
batch_norm
```
