### Uso practico de NumPy

Fórmula del error cuadrático medio, fundamental para los modelos de aprendizaje automático supervisado que abordan problemas de regresión

![](https://jalammar.github.io/images/numpy/mean-square-error-formula.png)

```python
n = 3
predictions = np.array([[1],[1],[1]])
labels = np.array([[1],[2],[3]])

error = (1/n) * np.sum(np.square(predictions - labels))
error
```
