### Essential Python Libraries
#### *NumPy*
- **NUM**erical **PY**thon : NumPy
- Se usa para trabajar con números y datos de forma rápida y eficiente en Python.

```python
# Sin numpy
a = [1,2,3]
b = [4,5,6]
res = [a[i] * b[i] for i in range(3)]
res
```

```python
# Con numpy
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

res = a * b
res
```

Short, faster and clear

#### **Pandas**

- Pandas es una librería de Python diseñada para trabajar con datos estructurados, como tablas (tipo Excel o bases de datos).
- Si NumPy es para números y matrices, Pandas es para datos del mundo real: CSVs, Excel, logs, datasets, etc.

```python
import pandas as pd

data = {
    "name":["Ana", "Luis", "Carlos"],
    "age":[23,25,29]
}

df = pd.DataFrame(data)
df
```

```python
# Real Example: filter by age
df_filter = df[df["age"] > 25]
df_filter
```

### **Matplotlib**
- Es la librería estándar de Python para crear gráficos.
- Sirve para visualizar datos: convertir números en imágenes que puedas entender rápido.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.title("Simple Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
